```python
import numpy as np
import matplotlib.pyplot as plt

# H and V values over time
deltaH = [-13, -167, -374, -370, -344, -199, -229, -382, -369]
deltaV = [-307, -159, 48, 40, 14, -129, -98, 58, 19]

N = len(deltaH)
T = 0.1  # Sample interval (seconds)

x = np.linspace(0.0, N*T, N)

# FFT for H
yf_H = np.fft.fft(deltaH)
xf_H = np.fft.fftfreq(N, T)[:N//2]
amplitudes_H = 2.0/N * np.abs(yf_H[0:N//2])

# FFT for V
yf_V = np.fft.fft(deltaV)
xf_V = np.fft.fftfreq(N, T)[:N//2]
amplitudes_V = 2.0/N * np.abs(yf_V[0:N//2])

# Plotting
plt.figure(figsize=(14, 6))

plt.subplot(2, 2, 1)
plt.plot(x, deltaH, marker='o', color='blue')
plt.title('H (Horizontal Difference) Over Time')
plt.xlabel('Time (s)')
plt.ylabel('H')

plt.subplot(2, 2, 2)
plt.stem(xf_H, amplitudes_H, basefmt=" ")
plt.title('Fourier Spectrum of H')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 3)
plt.plot(x, deltaV, marker='o', color='green')
plt.title('V (Vertical Difference) Over Time')
plt.xlabel('Time (s)')
plt.ylabel('V')

plt.subplot(2, 2, 4)
plt.stem(xf_V, amplitudes_V, basefmt=" ")
plt.title('Fourier Spectrum of V')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.grid(True)
plt.show()
```
